```python
import pandas as pd
whoiswho = pd.read_csv("cross-verified-database.csv", encoding='utf-8', encoding_errors='replace')
```

```python
whoiswho = whoiswho.sort_values(by='sum_visib_ln_5criteria',ascending=False)
```

```python
whoiswho = whoiswho.reset_index()
```

```python
whoiswho['no'] = whoiswho.index
```

## Kafka Among Other Notable People

```python
whoiswho.columns
```

```python
kafkas_importance = whoiswho[whoiswho['name'] == 'Franz_Kafka']['sum_visib_ln_5criteria'].max()
kafkas_importance
```

```python
len(whoiswho[whoiswho['sum_visib_ln_5criteria'] > kafkas_importance])
```

## Kafka Among Other Writers

```python
writing_jobs = ['writer','author','playwright','novelist','journalist']
whoiswho[(whoiswho['sum_visib_ln_5criteria'] > kafkas_importance) & (whoiswho['level3_main_occ'].isin(writing_jobs))][['name','no']]
```
