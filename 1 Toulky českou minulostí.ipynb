## 30 pod 30

Inspirováno jednou učitelkou dějepisu na TikToku.

```python
def thirty(year, area):
    whoiswho_filtered = whoiswho[whoiswho["area1_of_rattachment"].isin(area)]
    whoiswho_filtered = whoiswho_filtered[(whoiswho_filtered["birth"] > year - 30) & (whoiswho_filtered["birth"] < year - 17)]
    whoiswho_filtered = whoiswho_filtered.sort_values(by="sum_visib_ln_5criteria", ascending=False).head(30) 
    whoiswho_filtered["age"] = year - whoiswho_filtered["birth"]
    return whoiswho_filtered[["name", "age"]].set_index("name")
```

```python
thirty(1848, ceskezeme)
```

```python
thirty(1918, ceskezeme)
```

```python
thirty(1968, ceskezeme)
```
